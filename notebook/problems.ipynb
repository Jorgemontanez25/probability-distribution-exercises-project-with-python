## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO
from scipy.stats import norm 
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

#Define all the probability functions related to the normal distribution, for normal distribution the mean is by default equal 0 and the standard deviation by default is equal to 1
def dnorm(x, mean=0, std=1):
    """Density of the normal distribution."""
    return norm.pdf(x, loc=mean, scale=std)

def pnorm(x, mean=0, std=1):
    """Cumulative distribution function of the normal distribution."""
    return norm.cdf(x, loc=mean, scale=std)

def qnorm(p, mean=0, std=1):
    """Quantile function of the normal distribution."""
    return norm.ppf(p, loc=mean, scale=std)

def rnorm(n, mean=0, std=1):
    """Generate random variables from the normal distribution."""
    return norm.rvs(size=n, loc=mean, scale=std)
```

```python
#Set a random seed for reproducibility

np.random.seed(42)

#defining plotting function that takes data and corresponding PDF, CDF, and quantile values and plots them in a 3x1 subplot

def plot_distribution_functions(data, pmf, cdf, ppf):
    """Plot probability density function, cumulative distribution function, and quantile function."""
    fig, axes = plt.subplots(3, 1, figsize=(5, 7))

    axes[0].plot(data, pmf, "r-", label="PDF")
    axes[1].plot(data, cdf, "b-", label="CDF")
    axes[2].plot(data, ppf, "g-", label="Quantile Function")

    for ax, title in zip(axes, ["Probability Density Function", "Cumulative Distribution Function", "Quantile Function"]):
        ax.set_title(title)
        ax.legend()

    plt.tight_layout()
    plt.show()

#The main function where the normal distribution is defined, and the plotting function is called

def main():
    mean = 0
    std = 1
    data = np.arange(-5, 5, 0.01)

    pmf = norm.pdf(data, loc=mean, scale=std)
    cdf = norm.cdf(data, loc=mean, scale=std)
    ppf = norm.ppf(data, loc=mean, scale=std)

    plot_distribution_functions(data, pmf, cdf, ppf)

if __name__ == "__main__":
    main()
```

```python
#Define all the probability functions related to the uniform distribution, for uniform distribution where x, q and p are the parameters to evaluate 
#and the lower bound of the uniform distribution is set by default to 0 and the upper bound of the uniform distribution is set by default to 1

def dunif(x, low=0, high=1):
    """
    Calculates the probability density function of the uniform distribution.
    """
    return uniform.pdf(x, loc=low, scale=(high - low))

def punif(q, low=0, high=1):
    """
    Calculates the cumulative distribution function of the uniform distribution.
    """
    return uniform.cdf(q, loc=low, scale=(high - low))

def qunif(p, low=0, high=1):
    """
    Calculates the quantile function (inverse cumulative distribution) of the uniform distribution.
    """
    return uniform.ppf(p, loc=low, scale=(high - low))

def runif(n, low=0, high=1):
    """
    Generates random variables from the uniform distribution.
    """
    return uniform.rvs(loc=low, scale=(high - low), size=n)
```

```python
#defining plotting function that takes data and corresponding PDF, CDF, and quantile values and plots them in a 3x1 subplot

def plot_distribution_functions(data, pmf, cdf, ppf):
    """Plot probability mass function, cumulative distribution function, and quantile function."""
    fig, axes = plt.subplots(3, 1, figsize=(5, 7))

    axes[0].plot(data, pmf, "r-", label="PMF")
    axes[1].plot(data, cdf, "b-", label="CDF")
    axes[2].plot(data, ppf, "g-", label="Quantile Function")

    for ax, title in zip(axes, ["Probability Mass Function", "Cumulative Distribution Function", "Quantile Function"]):
        ax.set_title(title)
        ax.legend()

    plt.tight_layout()
    plt.show()

#The main function where the uniform distribution is defined, and the plotting function is called

def main():
    np.random.seed(42)

    low = 0
    high = 1
    data = np.arange(-5, 5, 0.01)

    pmf = dunif(data, low=low, high=high)
    cdf = punif(data, low=low, high=high)
    ppf = qunif(data, low=low, high=high)

    plot_distribution_functions(data, pmf, cdf, ppf)

if __name__ == "__main__":
    main()
```
